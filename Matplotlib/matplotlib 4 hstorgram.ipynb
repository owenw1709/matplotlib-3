```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print (matplotlib.__version__)
print (np.__version__)
```

Sample Data

```python
mu, sigma =100,15 #nilai mean dan nilai standard deviation

x =mu +sigma*np.random.randn(1000)# normal distribution
x
```

```python
x.shape
```

### Histrogram Dengan Pyplot Style

```python
plt.hist(x,
        bins =50,
        facecolor ='g',
        alpha=0.50)

plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Contoh Histogram')

plt.text(45,500,r'$\mu=100, \ \ sigma=15$')
plt.grid()

plt.show()
```

### Histrogram Dengan OO Style

```python
fig,ax =plt.subplots()

ax.hist(x,
        bins =50,
        facecolor ='r',
        alpha=0.75)

ax.set_xlabel('Sumbu X')
ax.set_ylabel('Sumbu Y')
ax.set_title('Contoh Histogram')

ax.text(45,500,r'$\mu=100, \ \ sigma=15$')
ax.grid()

plt.show()
```
